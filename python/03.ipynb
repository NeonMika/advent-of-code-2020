```python
def countSlope(data, x_step, y_step):
    row = 0
    col = 0
    
    width = len(data[0])
    last_row = len(data) - 1

    hits = 0

    while(row != last_row):
        row = row + y_step
        col = (col + x_step) % width
        hits = hits + data[row][col]
        #print(row, col, hits)
    
    #print(hits)
    return hits
```
